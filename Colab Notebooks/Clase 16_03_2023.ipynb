```python
#ejercicio 9

puntuacion = float(input("Cual ha sido la puntuacion:"))

if puntuacion == 0.0:
  print("Su nivel es inaceptable y ha recibido 0€")
elif puntuacion == 0.4:
  print("Su nivel es aceptable y ha recibido " + 0.4*2400 + "€")
elif puntuacion >= 0.6:
  print("su nivel es meritorio y ha recibido " + str(puntuacion)*2400 + "€")
else:
  print("error")
```
