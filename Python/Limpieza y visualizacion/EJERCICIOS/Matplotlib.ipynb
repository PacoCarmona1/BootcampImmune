## Ejercicio 1: lee el beneficio total de todos los meses y muéstralo mediante un diagrama de líneas

Datos de ganancias totales proporcionados para cada mes. El diagrama de líneas generado debe incluir las siguientes propiedades: –

    Nombre de etiqueta X = Número de mes
    Nombre de la etiqueta Y = Beneficio total

El gráfico de trazado de líneas debería tener este aspecto.

```python
import pandas as pd
datos_ventas = pd.read_csv("/Users/Paco/Downloads/ventas.csv")
datos_ventas
```

```python
import matplotlib.pyplot as plt

# Obtener los datos de número de mes y beneficio total
meses = datos_ventas['month_number']
beneficio_total = datos_ventas['total_profit']

# Crear el gráfico de líneas
plt.plot(meses, beneficio_total, marker='o', linestyle='-', color='b')

# Configurar las etiquetas de los ejes
plt.xlabel('Número de mes')
plt.ylabel('Beneficio total')

# Título del gráfico
plt.title('Beneficio total por mes')

# Mostrar el gráfico
plt.show()
```

## Ejercicio 2: obten el beneficio total de todos los meses y muestra el diagrama de líneas con las siguientes propiedades de estilo

El diagrama de líneas generado debe incluir las siguientes propiedades de estilo: –

## Ejercicio 3: lee todos los datos de ventas de productos y muéstralos usando un gráfico multilínea

Muestre el número de unidades vendidas por mes para cada producto utilizando gráficos de líneas múltiples. (es decir, trama separada para cada producto).

El gráfico debería verse así.

```python
# Obtener los datos de unidades vendidas por mes para cada producto
meses = datos_ventas['month_number']
productos = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Configurar el gráfico de líneas múltiples
plt.figure(figsize=(7, 5))  # Ajustar el tamaño del gráfico según sea necesario

# Iterar sobre cada producto y trazar el gráfico de líneas correspondiente
for producto in productos:
    unidades_vendidas = datos_ventas[producto]
    plt.plot(meses, unidades_vendidas, label=producto, marker='o', lw=3)

# Etiquetas de los ejes
plt.xlabel('Número de mes')
plt.ylabel('Número de unidades vendidas')

# Leyenda en la ubicación superior izquierda
plt.legend(loc='upper left')

# Mostrar el gráfico
plt.show()
```

## Ejercicio 4: Lee los datos de ventas de pasta de dientes de cada mes y muéstralo usando un diagrama de dispersión

Además, agregue una cuadrícula en la trama. el estilo de línea de cuadrícula debe ser “–“.

## Ejercicio 5: Lee los datos de ventas de productos de cremas y lavados faciales y muéstralos usando el gráfico de barras

El gráfico de barras debe mostrar el número de unidades vendidas por mes para cada producto. Agregue una barra separada para cada producto en el mismo gráfico.

El gráfico de barras debería verse así.

```python
import numpy as np

meses = datos_ventas['month_number']
cremas = datos_ventas['facecream']
lavados_faciales = datos_ventas['facewash']

# Configurar los índices de posición de las barras
indices = np.arange(len(meses))

# Configurar el ancho de las barras
ancho_barras = 0.35

# Configurar el gráfico de barras
plt.bar(indices, cremas, width=ancho_barras, label='Cremas')
plt.bar(indices + ancho_barras, lavados_faciales, width=ancho_barras, label='Lavados faciales')

# Etiquetas de los ejes
plt.xlabel('Número de mes')
plt.ylabel('Número de unidades vendidas')

# Ticks del eje x
plt.xticks(indices + ancho_barras / 2, meses)

# Leyenda en la ubicación superior derecha
plt.grid(linestyle='--', linewidth=1)
plt.legend(loc='upper right')

# Mostrar el gráfico de barras
plt.show()
```

## Ejercicio 7: Lee el beneficio total de cada mes y muéstralo usando el histograma para ver los rangos de beneficio más comunes

El histograma debería verse así.

```python
fig = plt.figure(figsize=(4.5, 3))

# Creamos las variables
total_profit = datos_ventas['total_profit'].tolist()

# Creamos los contenedores donde se filtrarán los datos
contenedores = [175000, 200000, 225000, 250000, 300000, 350000]

# Crear el histograma
plt.hist(total_profit, bins=contenedores, label='Profit Data')

# Configurar etiquetas y título
plt.xlabel('Profit Range in Dollar')
plt.ylabel('Actual profit in Dollar')
plt.title('Profit Data')

# Configurar ticks del eje x
plt.xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000], rotation=45)

# Delimitar el rango del eje x
plt.xlim(130000, 360000)

# Asignar posición y tamaño a la leyenda
plt.legend(loc='upper left', prop={'size': 8})

# Guardar el diagrama en disco duro
plt.savefig('histograma_profit.png')

# Mostrar el histograma
plt.show()
```

## Ejercicio 8: Calcula los datos de ventas totales del último año para cada producto y muéstralos usando un gráfico circular (pie chart)

Nota: En la visualización del gráfico circular Número de unidades vendidas por año para cada producto en porcentaje.

El gráfico circular debería verse así.

```python
# Obtener los datos de ventas totales del último año para cada producto
ventas_ultimo_año = datos_ventas.iloc[-12:][['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
productos = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

# Calcular el total de unidades vendidas por producto en el último año
ventas_totales = ventas_ultimo_año.sum()

# Configurar el gráfico circular (pie chart)
plt.pie(ventas_totales, labels=productos, autopct='%1.1f%%')

# Título del gráfico
plt.title('Unidades vendidas por producto en el último año')

# Agregar la leyenda
plt.legend(loc='lower right')

# Mostrar el gráfico circular
plt.show()
```

## Ejercicio 9: Lee el lavado de cara de jabón de baño de todos los meses y muéstrelo usando la trama secundaria

```python
# Obtener los datos de lavado de cara y jabón de baño de todos los meses
lavado_cara = datos_ventas['facewash']
jabon_baño = datos_ventas['bathingsoap']

# Crear la figura y los ejes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7,5), sharex=True)

# Graficar el lavado de cara en la primera gráfica
ax1.plot(meses, lavado_cara, color='black', marker='o', lw=3)
ax1.set_ylabel('Lavado de Cara')

# Graficar el jabón de baño en la segunda gráfica
ax2.plot(meses, jabon_baño, color='red',marker='o', lw=3)
ax2.set_ylabel('Jabón de Baño')
ax2.set_xlabel('Número de mes')

# Ajustar los espacios entre las gráficas
plt.tight_layout()

# Mostrar las gráficas
plt.show()


```

## Ejercicio 10: Lee todos los datos de ventas de productos y muéstralos usando el diagrama de pila

El diagrama de pila debería verse así.

```python
# Obtener los datos de ventas de cada producto
facecream = datos_ventas['facecream']
facewash = datos_ventas['facewash']
toothpaste = datos_ventas['toothpaste']
bathingsoap = datos_ventas['bathingsoap']
shampoo = datos_ventas['shampoo']
moisturizer = datos_ventas['moisturizer']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 5))


# Graficar los datos de ventas de cada producto en el diagrama de pila con los colores definidos
ax.stackplot(meses, facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer,
             labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'])

# Configurar etiquetas y título
ax.set_xlabel('Número de mes')
ax.set_ylabel('Unidades vendidas')
ax.set_title('Ventas de productos')

# Ubicar la leyenda en la parte superior izquierda
ax.legend(loc='upper left')

# Mostrar la figura
plt.show()
```
