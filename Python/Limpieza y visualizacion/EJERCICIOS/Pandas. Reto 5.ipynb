### 1. Generar un DataFrame con los datos de los cuatro ficheros.

```python
import pandas as pd

# Lista de nombres de los archivos
archivos = ['/Users/Paco/Desktop/emisiones/emisiones-2016.csv', '/Users/Paco/Desktop/emisiones/emisiones-2017.csv', '/Users/Paco/Desktop/emisiones/emisiones-2018.csv', '/Users/Paco/Desktop/emisiones/emisiones-2019.csv']

# Lista para almacenar los DataFrames de cada archivo
dataframes = []


# Leer cada archivo CSV con separaci√≥n por punto y coma (;) y a√±adirlo al DataFrame
for archivo in archivos:
    df = pd.read_csv(archivo, sep=';')
    dataframes.append(df)

# Concatenar los DataFrames en uno solo
df_final = pd.concat(dataframes)

# Mostrar el DataFrame resultante
df_final
```

### 2. Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, A√ëO, MES y las correspondientes a los d√≠as D01, D02, etc.

```python
# Filtrar las columnas requeridas
# Se utiliza la funcion 'startwith' para indentificar las columnas de los dias y luego se a√±ade a las columnas requeridas 
columnas_dias = [col for col in df_final.columns if col.startswith('D')]
columnas_requeridas = ['ESTACION', 'MAGNITUD', 'ANO', 'MES'] + columnas_dias
df_filtrado = df_final.filter(items=columnas_requeridas)

# Mostrar el DataFrame resultante
df_filtrado
```

### 3. Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los d√≠as aparezcan en una √∫nica columna.

```python
# Utilizamos la funci√≥n melt() para reestructurar el DataFrame. Esto permite crear una unica columna DIA y VALOR (var_name y value_name).
df_reestructurado = df_filtrado.melt(id_vars=['ESTACION', 'MAGNITUD', 'ANO', 'MES'],
                                     var_name='DIA',
                                     value_name='VALOR')

# Mostrar el DataFrame resultante
df_reestructurado
```

### 7. Crear una funci√≥n que reciba una estaci√≥n, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estaci√≥n y rango de fechas dado.

```python
# Crea una fundion
def obtener_emisiones(estacion, contaminante, fecha_inicio, fecha_fin):
    filtro = (df_reestructurado['ESTACION'] == estacion) & \
             (df_reestructurado['MAGNITUD'] == contaminante) & \
             (df_reestructurado['FECHA'].between(fecha_inicio, fecha_fin))
    emisiones = df_reestructurado.loc[filtro, 'VALOR']
    return emisiones

```

Se pedir√° que ingreses el n√∫mero de la estaci√≥n, el n√∫mero del contaminante, la fecha de inicio y la fecha de fin. Aseg√∫rate de ingresar los valores correctamente en el formato solicitado.

```python
def emisiones_medias_mensuales(df, contaminante, a√±o):
    df_filtrado = df[(df['MAGNITUD'] == contaminante) & (df['ANO'] == a√±o)]
    emisiones_medias = df_filtrado.groupby(['ESTACION', 'MES'])['VALOR'].mean()
    return emisiones_medias

contaminante = int(input("Ingrese el n√∫mero de contaminante: "))
a√±o = int(input("Ingrese el a√±o: "))

# Obtener las emisiones medias mensuales
emisiones_medias = emisiones_medias_mensuales(df_reestructurado2, contaminante, a√±o)
emisiones_medias
```

### 12. Crear un funci√≥n que reciba una estaci√≥n de medici√≥n y devuelva un DataFrame con las medias mensuales de los distintos tipos de contaminantes.

```python
#Muestrame las diferentes estaciones

estaciones = df_reestructurado2['ESTACION'].unique()
print(estaciones)
```

```python
#Crear la funcion 

def medias_mensuales_contaminantes(df, estacion):
    df_estacion = df[df['ESTACION'] == estacion]
    medias_mensuales = df_estacion.groupby(['MAGNITUD', 'MES'])['VALOR'].mean()
    df_medias_mensuales = medias_mensuales.unstack(level=0)
    return df_medias_mensuales

#Imput para ingresar la estacion
estacion = int(input("Ingrese el n√∫mero de estaci√≥n: "))

# Obtener el DataFrame con las medias mensuales de los contaminantes
df_medias_mensuales = medias_mensuales_contaminantes(df_reestructurado2, estacion)
print(df_medias_mensuales)
```
