```python
# Agrupaciones, usamos groupby y groups para agrupar por categorias segun una serie de 

bebidas =  pd.read_csv('/Users/Paco/Downloads/bebidas.csv')
bebidas
```

```python
# ver la media de porciones de cervezas mundial

bebidas.porciones_cerveza.mean()
```
