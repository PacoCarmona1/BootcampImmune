```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

# Distribucion en Seaborn

Analizaremso los graficos que nos permiten visualozar la distribucion de un conjunto de datos. Paquetes:


- dispot
- joinplot
- pairplot
- rugpllot

```python
tips = sns.load_dataset("tips")
tips
```

# 1. Diagramas de distribucion

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=20)
```

# Diagramas de conjuto

jointplot permite hacer coincidir dos distplot para dato variado. Con el parametro kind podemos elegir como podemos mmostrar estas comaprativas.

* scatter 
* reg
* resid
* kde
* hex

```python
sns.jointplot(x ='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x ='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x ='total_bill', y='tip', data=tips, kind='kde')
```

# Graficos de pares

pairplot: trazar relaciones por pares en un marco de datos completos (columnas numericas) y admite tonos de clÂ¡olor para columnas ccategoricas

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

#  Diagrama de alfombra

rugplot: Esta funcion dibuja un quion para cada punto en una distribucion univariada. esto nos va a servir como bloque de costruccion para los diagramas KDE(kernel Density Estimation)

```python
sns.rugplot(tips['total_bill'])
```

# Diagramas de Matriz

Los graficos matriciales permiten trazar datos con matrices codificadas cpor colores y tambien se pueden usar par indicar grupos detro de los datos

```python
flight = sns.load_dataset('flights')
flight
```

```python
# quiero analizar la relacion entre variables

correlacion = tips.corr()
correlacion
```

```python
correlacion2 = flight.corr()
correlacion2
```

# Mapas de calor

heatmap: para que un mapa de calor funcione necesitamos que los datos tienen que estar en una matriz

```python
# Quiero analizar mis relaciones entre variables de manera visual. Uso un color para representar 

sns.heatmap(correlacion2)
```

```python
sns.heatmap(correlacion, annot=True)
```

```python
sns.heatmap(correlacion, annot=True, cmap='coolwarm')
```

```python
# quiero hacer un analisis de dsitribucion de los datos de la aerolinea.

flight.head()
```

```python
# Necesito que mis datos esten en formato matriz

df1 = flight.pivot(index='year', columns='month', values='passengers')
df1
```

```python
sns.heatmap(df1)
```

```python
df2 = flight.pivot(index='month', columns='year', values='passengers')
df2
```

```python
sns.heatmap(df1, cmap='rainbow', linewidths=5)
```

# Mapas de cluster

clustermap: utiliza la agrupacio jerarquica para producir una version agrupada del mapa de calor. Agrupa las columnas que son parecidas las unas de las otras.

```python
sns.clustermap(df2)
```

```python
sns.clustermap(df2, standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
iris_norm = iris.iloc[:, 0:4].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
iris_norm
```

```python
sns.clustermap(iris_norm, standard_scale=1)
```

# Graficos de regresion

lmplot: permite mostrar modelos lineales, pero tambien permite dividri estos graficos en funcion de las caracteristicas, asi como categorizar por color.

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, row='sex')
```

```python
male_tips = tips[tips['sex'] == 'Male']
sns.lmplot(x='total_bill', y='tip', data=male_tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex',palette='coolwarm')
```
