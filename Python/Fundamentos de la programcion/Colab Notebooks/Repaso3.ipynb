Ejercicio 12:

Crear una funcion lista_suma_mayor(x, y) que reciba dos parametros de tipo lista y devuelva la lista cuya suma de sus valores sea mayor

*Ejemplo de input y output:*

*Input: x=[2,1,3,5], y=[4,0,1,0]

*Output: [2,1,3,5]*

*Explicacion: la suma de los valores de x es 2+1+3+5=11 y la suma de los valores de y es 4+0+1+0=5 por lo tanto devuelvo la lista x porque la suma de sus valores es 11 que es mayor a la suma de los valores de la suma de la lista y que es 5

```python
def lista_suma_mayor(x,y):
   #TODO: escribir aqui el codigo de su solucion
   suma_x = 0
   suma_y = 0
    
   for i in range(len(x)):
        suma_x += x[i]
        
   for i in range(len(y)):
        suma_y += y[i]
        
   if suma_x > suma_y:
        return x
   else:
        return y
```

```python
#NO MODIFICAR EL SIGUIENTE CODIGO
#Codigo de test para vefiricar si la funcion es correcta
assert lista_suma_mayor([2,1,3,5], [4,0,1,0]) == [2,1,3,5], "Error en el test 1"
assert lista_suma_mayor([2], [3,6]) == [3,6], "Error en el test 2"
assert lista_suma_mayor([],[]) == [], "Error en el test 3"
assert lista_suma_mayor([1,2,3],[9,8,7]) == [9,8,7], "Error en el test 4"
print("Felicitaciones! Todos los tests fueron pasados exitosamente!")
```

Crear una funcion sumar_listas2(x, y) que reciba dos parametros de tipo lista y devuelva una nueva lista con la suma de los valores de la primera mas los de la segunda invertida, es decir, lista_nueva[i] = x[i] + y[n-1-i] donde n = len(x)

Aclaraciones:

- x e y siempre tendran la misma cantidad de elementos

*Ejemplo de input y output:*

*Input: x=[2,1,3,5], y=[4,0,1,0]*

*Output: [2,2,3,9]*

```python
def sumar_listas2(x,y):
  #TODO: escribir aqui el codigo de su solucion
  n = len(x)
  resultado = []
  for i in range(n):
     resultado.append(x[i] + y[n-1-i])
  return resultado
```

```python
#NO MODIFICAR EL SIGUIENTE CODIGO
#Codigo de test para vefiricar si la funcion es correcta
assert sumar_listas2([2,1,3,5], [4,0,1,0]) == [2,2,3,9], "Error en el test 1"
assert sumar_listas2([2], [3]) == [5], "Error en el test 2"
assert sumar_listas2([],[]) == [], "Error en el test 3"
assert sumar_listas2([1,2,3],[9,8,7]) == [8,10,12], "Error en el test 4"
print("Felicitaciones! Todos los tests fueron pasados exitosamente!")
```

Ejercicio 14:

Crear una funcion suma_de_aleatorios(semilla) que reciba un parametro de tipo int y genere numero aleatorios utilizando la funcion random.randrange(0,100) hasta que se genere el numero cero. Devolver la suma de todos los numeros generados.

*Ejemplo de input y output:*

*Input: semilla=1*

*Output: 1128*

```python
import random
def suma_de_aleatorios(semilla):
  random.seed(semilla)
  #TODO: escribir aqui el codigo de su solucion
  num = random.randrange(0, 100)
  suma = 0
  while num != 0:
    suma += num
    num = random.randrange(0, 100)
  return suma
```

```python
#NO MODIFICAR EL SIGUIENTE CODIGO
#Codigo de test para vefiricar si la funcion es correcta
assert suma_de_aleatorios(1) == 1128, "Error en el test 1"
assert suma_de_aleatorios(2) == 8319, "Error en el test 2"
assert suma_de_aleatorios(3) == 6683, "Error en el test 3"
assert suma_de_aleatorios(42) == 1693, "Error en el test 4"
print("Felicitaciones! Todos los tests fueron pasados exitosamente!")
```
