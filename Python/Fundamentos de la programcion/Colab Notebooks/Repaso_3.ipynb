### Como iterar una lista (recorrer sus elementos uno a uno)

```python
mi_lista = [10,20,30,40,50]
```

Utilizando el bucle for (recomendado)

```python
#utilizando un for sobre la lista
#en este caso solo necesito saber 
#el valor de los elementos y no de los indices
for elemento in mi_lista:
  print("Elemento:", elemento)
```

```python
#utilizando un for sobre los indices de la lista
#en este caso necesito saber 
#el valor de los elementos y de los indices
for i in range(len(mi_lista)):
  print("Indice:", i, "Elemento:", mi_lista[i])
```

```python
#otra opcion cuando necesito los indices y los elementos
for i, elemento in enumerate(mi_lista):
  print("Indice:", i, "Elemento:", elemento)
```

Utilizando el bucle while

```python
i = 0
while i < len(mi_lista):
  print("Indice:", i, "Elemento:", mi_lista[i])
  i += 1
```

### Funciones

Crear una funcion

```python
def decir_hola():
  print("Hola soy una funcion")
```

Llamar a una funcion

```python
decir_hola()
```

Funciones que reciben variables como parametros

```python
def decir_hola(nombre):
  print("Hola", nombre, "soy una funcion")
```

```python
decir_hola("Maria")
```

```python
def decir_hola(nombre, veces):
  for _ in range(veces):
     print("Hola", nombre, "soy una funcion")
```

```python
decir_hola("Juana", 3)
```

Funciones que devuelven un valor

```python
def uno_menos(numero):
  x = numero - 1
  return x
```

```python
uno_menos(10)
```

```python
def todo_mayusculas_menos_primera_letra(texto):
  s = texto[0].lower() + texto[1:].upper()
  return s;
```

```python
todo_mayusculas_menos_primera_letra("Helicoptero")
```

```python
def convertir_string_a_lista(texto):
  ret = []
  for caracter in texto:
    ret.append(caracter)
  return ret
```

```python
convertir_string_a_lista("El cielo es azul")
```
