```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# queremo ver ls distribucion de los datos para analiza el coste del billete y la edad

sns.jointplot(x ='age', y='fare', data=titanic)
```

```python
# Ahora quiero expandir mi distribucion de coste de billete, lo quiero en rojo y con 40 cajas y medio transparente


sns.displot(titanic['fare'], bins=40, color='red', alpha=0.5, kde=True)
```

```python
# Quiero hacer un perfilado de datos para empezar con mis manipullaciones, edad y clase

sns.boxplot(x='class', y='age', data=titanic, palette='rainbow')
```
