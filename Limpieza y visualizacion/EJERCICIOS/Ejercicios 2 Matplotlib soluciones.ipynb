```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('\Users/Paco/Downloads/vgsales.csv')
```

```python
# Pregunta 1: Plataforma más popular
plt.figure(figsize=(10,6))
df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Ventas globales por plataforma')
plt.ylabel('Ventas globales (en millones)')
plt.show()
```

```python
# Pregunta 2: Ventas a lo largo del tiempo
plt.figure(figsize=(10,6))
df.groupby('Year')['Global_Sales'].sum().plot(kind='line')
plt.title('Ventas globales por año')
plt.ylabel('Ventas globales (en millones)')
plt.show()
```

```python
# Pregunta 3: Géneros más populares por región
regions = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
for region in regions:
    plt.figure(figsize=(10,6))
    df.groupby('Genre')[region].sum().sort_values(ascending=False).plot(kind='bar')
    plt.title(f'Ventas por género en {region}')
    plt.ylabel('Ventas (en millones)')
    plt.show()
```

```python
# Pregunta 4: Editores más exitosos
plt.figure(figsize=(10,6))
df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Ventas globales por editor')
plt.ylabel('Ventas globales (en millones)')
plt.show()
```
