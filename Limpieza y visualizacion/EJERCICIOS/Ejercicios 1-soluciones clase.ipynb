## Ejercicio 3

**Crea la siguiente gráfica agregando dos ejes a un objeto de figura en [0,0,1,1] y [0.2,0.5,.4,.4]**

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])
```

**Ahora usa las matrices x, y y z para recrear el diagrama a continuación. Observa los límites x y y en el gráfico insertado(set_xlim):**

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z)
axes2.plot(x,y)

axes1.set_xlabel('x')
axes1.set_ylabel('z')

axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom')

axes2.set_xlim(20,22)
axes2.set_ylim(30,50)

fig;
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z)
axes2.plot(x,y)

axes1.set_xlabel('x')
axes1.set_ylabel('z')

axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('zoom')

axes2.set_xlim([20.0,22.0])
axes2.set_ylim([30,50])

axes1.axis('off')

fig
```

## Ejercicio 4

**Usa plt.subplots(nrows=1, ncols=2) para crear el diagrama a continuación.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

**Ahora traza (x,y) y (x,z) en los ejes. Juega con el ancho de línea y el estilo**

```python
axes[0].plot(x,y,color='blue',ls='--', lw=3)
axes[1].plot(x,z,color='red',ls='-', lw=3)
fig
```

**Mira si puedes cambiar el tamaño de la gráfico agregando el argumento figsize() en plt.subplots()  copiando y pegando el código anterior.**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,2))

axes[0].plot(x,y,color="blue", lw=5)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z,color="red", lw=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
```
