```python
import seaborn as sns
import numpy as np
```

# 1. Graficos de datos categóricos

Ahora analicemos el uso de seaborn para trazar datos categoricos. Hay algunos tipos de tramas principales para esto:

-factorplot
-boxplot
-violinplot


```python
tips = sns.load_dataset('tips')
tips
```

# 2. Diagramas de barras y conteo

Para pintar datos categoricos y lo usamo como hicimos en matplotlib (usando la funcion barplot)m

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# Por defecto el argumento estimador nos calcula la media, vamos a modificarlo:


sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)

# cargamos la libreria de numpy para usar la funcion np.std
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std, ci=95)
```

# 3. Diagrama de conteo

Countplot: es lo mismo que el diagrama de barras solo que el estiamdor cuenta explicacitamente el numero de ocurrencia

```python
sns.countplot(x='sex', data=tips)
```

# 4. Boxplots y violin plots

Dispersion y distribucion del conjunto de datos

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow' )
```

```python
# Analiza a mis clientes y sus habitos

sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='smoker')
```

```python
smoker_yes = tips[tips['smoker'] == 'Yes']
sns.boxplot(x='day', y='total_bill', data=smoker_yes, palette='rainbow')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

# 5. Violinplot

juega un papel similar al de boxplot, muestra valores de distribucion de datos cuantitativos a traves de varios niveles de uno(o varios) variables categoricas

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow' )
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='Set2', hue='sex', split=True)
```

# Diagramas de emjabre o de franjas

Stripplot: dibuja un diagrama de dispersion donde una variable es la categorica.
Esto se podria usar como complemento para otras visualizaciones

```python
sns.stripplot(x='day', y='total_bill', data=tips, palette='Set1')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, palette='Set1', jitter=False)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, palette='Set1', hue='sex', dodge=True)
```

# 6. Swarmplot 

Se asimila al anterior pero los puntos se ajustan (solo a lo largo del eje categorico) para que no se superpongan.

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow')
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```
