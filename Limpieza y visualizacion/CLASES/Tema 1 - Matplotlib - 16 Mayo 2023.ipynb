```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

## Comandos b√°sicos de Matplotlib

```python
# Primero especificamos los datos que queremos pintar

import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# Ahora pintamos los datos en un grafico de lineas rojo

plt.plot(x,y,'r')
plt.xlabel('Etiqueta eje X')
plt.ylabel('Etiqueta eje Y')
plt.title('Titulo del grafico')
plt.show()
```

### Creaci√≥n de Multiplots en un mismo lienzo

```python
# plt.subplot(numero de filas, numero de columnas, grafico en cuestion)
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

## Metodo orientado a objetos

```python
fig= plt.figure()

# a√±adir ejes a las figuras
axes=fig.add_axes([0.1,0.1,0.8,0.8])

# introducimos los datos y etiquetas de ejes

axes.plot(x,y,'b')
axes.set_xlabel('Etiqueta eje X')
axes.set_ylabel('Etiqueta eje Y')
axes.set_title('Titulo del grafico')
```

```python
# creemos un lienzo en blanco para crear un grafico dentro de un grafico

fig = plt.figure()

axes1= fig.add_axes([0.1,0.1,0.8,0.8]) # eje principal
axes2= fig.add_axes([0.2,0.5,0.4,0.3]) # eje inset

# Datos para mi grafico principal

axes1.plot(x,y,'b')
axes1.set_xlabel('Etiqueta X')
axes1.set_ylabel('Etiqueta Y')
axes1.set_title('Titulo principal')

# datos para el eje secundario

axes2.plot(y,x,'r')
axes2.set_xlabel('Etiqueta X')
axes2.set_ylabel('Etiqueta Y')
axes2.set_title('Titulo secundario')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# Ahora iterando 

for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('Etiqueta X')
    ax.set_ylabel('Etiqueta Y')
    ax.set_title('Titulo secundario')
    
fig
```

```python
### DPI, tama√±o de la figura, ratio de aspecto

fig = plt.figure(figsize=(8,4),dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(1,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

### Leyendas

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label='x**2')
ax.plot(x,x**3,label='x**3')
ax.legend()
```

```python
ax.legend(loc=4)
fig
```

```python
# mas colores y estilos de linea

fig, ax=plt.subplots()

ax.plot(x,x**2,'b.-')
ax.plot(x,x**3,'g--')
```

```python
# colores con el parametro color

fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # alpha es la transparencia
ax.plot(x,x+2, color= "#8B008B")
ax.plot(x,x+3, color= "#FF8C00")
```

```python
# Ahora probemos a visualizar diferentes resultados de grafico de lineas para cada funcion:

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# ESTILOS DE LINEA ‚Äò-‚Äò, ‚Äò‚Äì‚Äô, ‚Äò-.‚Äô, ‚Äò:‚Äô, ‚Äòsteps‚Äô
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# MARCADOR CUSTOMIZADO
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# ESTILOS DE MARCADOR = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# TAMA√ëO DEL MARCADOR Y COLOR
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
